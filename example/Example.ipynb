### Imports

```python
import pandas as pd
from aalogo import AALogorizer, StandardConfig
```

### Import File as pandas.DataFrame

```python
# directory management
path, sep = StandardConfig.find_folderpath()
path, sep
```

```python
# import of test data sequences for visualization
test_data = pd.read_excel(f"DeepTMHMM_sub.xlsx")
test_data.head(5)
```

#### important columns:
    start_pos_TMD
    stop_pos_TMD
    sequence

## How it works:

```python
AALogorizer.AAlogoMaker.help()
```

```python
new_example = AALogorizer.AAlogoMaker(test_data, "Sub_DeepTMHMM", "sequence", "start_pos_TMD", "stop_pos_TMD")
                                        # df           name       column_seq          *args_position
```

```python
new_example.tmd_mode(config_name="OG_AA_config")
```

```python
new_example.tmd_mode.available_themes
```

```python
new_example.tmd_mode(theme="Cornette", aa_jmd=4, aa_tmd=10, headers=["JMD", "TMD"])
```

## Robust, even if sequences do not have the same size, both on the N and C terminal side

```python
new_example2 = AALogorizer.AAlogoMaker(test_data, "Sub_DeepTMHMM", "sequence", "start_pos_ecto")
```

```python
new_example2.single_mode(theme="Eisenberg", aa_right=4, aa_left=30, headers=["", "Ecto Domain"], custom_colors=[[206, 222, 151], [201, 56, 42]])
```
