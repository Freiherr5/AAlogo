```python
import pandas as pd
import AALogorizer
import StandardConfig
```

```python
data_expert = pd.read_excel("TMD_expert_curated_subtrates_nonsubstrates_FK.xlsx").set_index("entry")[["tmd", "name"]]
list_expert_entries = data_expert.index.tolist()
len(list_expert_entries)
data_expert
```

```python
data_deepTMHMM = pd.read_excel("example_data/DeepTMHMM_sub.xlsx").set_index("entry")[["sequence", "name", "start_pos_TMD", "stop_pos_TMD"]].loc[list_expert_entries]
data_deepTMHMM
```

```python
list_for_concat = []
for (e_index, e_rows), (index, rows) in zip(data_expert.iterrows(), data_deepTMHMM.iterrows()):
    start_pos = rows[0].find(e_rows[0]) + 1
    stop_pos = rows[0].find(e_rows[0]) + len(e_rows[0])
    list_for_concat.append([e_index, start_pos, stop_pos])
data_concat = pd.DataFrame(list_for_concat, columns=["entry", "start", "stop"]).set_index("entry")
data_visualize = pd.concat([data_deepTMHMM, data_concat], axis=1)
data_visualize
```

```python
expert_visualize = AALogorizer.AAlogoMaker(data_visualize, "Sublit_expert", "sequence", "start", "stop")
expert_visualize.tmd_mode(aa_jmd=4, aa_tmd=10, headers=["JMD", "TMD"])
```

```python
deepTMHMM_visualize = AALogorizer.AAlogoMaker(data_visualize, "Sublit_DeepTMHMM", "sequence", "start_pos_TMD", "stop_pos_TMD")
deepTMHMM_visualize.tmd_mode(aa_jmd=4, aa_tmd=10, headers=["JMD", "TMD"])
```
